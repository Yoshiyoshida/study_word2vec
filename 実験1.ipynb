#  国-首都の関係ベクトルを任意の首都ベクトルに足したらその国を出力できるか

```python
import gensim

model = gensim.models.Word2Vec.load("latest-ja-word2vec-gensim-model/word2vec.gensim.model")
```

```python
# 国と首都のリストを作成

con_list = []

f = open('country.txt', 'r')

datalist = f.readlines()
for i in range(193):
    # print(datalist[i].split("\t"))
    sample = datalist[i].split("\t")
    
    sample[1] = sample[1].rstrip()
    con_list.append(sample)
    
f.close()
```

```python
# 

for llist in con_list:
    print(llist)
    contury = str(llist[0])
    capital = str(llist[1])

    
    # 単語の意味を足し引きして、類似度の高いものを上から１０個表示する
    try:
        print(capital)
        print(contury)
        print(model.wv[capital])
        print(model.wv.most_similar(positive=[capital], negative=[contury]))
        print("　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　")
    except:
        continue
```

```python
# 国ベクトル - 首都ベクトル
relation_vec_list = []

for llist in con_list:
    print(llist)
    contury_name = str(llist[0])
    capital_name = str(llist[1])
    
    
    # 単語の意味を足し引きして、類似度の高いものを上から１０個表示する
    try:
        contury_vec = model.wv[contury_name]
        capital_vec = model.wv[capital_name]
        
        relation_vec = contury_vec - capital_vec
        relation_vec_list.append(relation_vec)

    except:
        continue
```

```python
len(relation_vec_list)
```

```python
# 平均関係ベクトル

sum_v = 0

for v in relation_vec_list:
    
    sum_v = v + sum_v

average_relation_vec = sum_v/len(relation_vec_list)
```

```python
top_list = []
count = 0
c_sum = 0
ok_list = []

for l in con_list:
    country_name = l[0]
    capital_name = l[1]
    
    try:
        capital_vec = model.wv[capital_name]
        may_country_vec = capital_vec + average_relation_vec
    
        most_similar = model.wv.most_similar([may_country_vec], [], 10)
        
        top_list = most_similar
        
        for kouho_capital in top_list:
            if kouho_capital[0] == capital_name:
                print("一致")
                print(kouho_capital[0])
                count += 1
                ok_list.append(country_name)
                break
    
        print(country_name + " : " + capital_name)
        print(most_similar)
        print("\n")
        c_sum += 1

    except:
        print("エラー")
        print(country_name + " : " + capital_name)
        continue
print((count/c_sum)*100)
```

# 平均関係ベクトルではなく、任意の関係ベクトルで推定できるか確認する

```python
import gensim

model = gensim.models.Word2Vec.load("latest-ja-word2vec-gensim-model/word2vec.gensim.model")
```

```python
# 国と首都のリストを作成

con_list = []

f = open('country.txt', 'r')

datalist = f.readlines()
for i in range(193):
    # print(datalist[i].split("\t"))
    sample = datalist[i].split("\t")
    
    sample[1] = sample[1].rstrip()
    con_list.append(sample)
    
f.close()
```

```python
# 国ベクトル - 首都ベクトル
relation_vec_list = []

for llist in con_list:
    contury_name = str(llist[0])
    capital_name = str(llist[1])
    
    
    # 単語の意味を足し引きして、類似度の高いものを上から１０個表示する
    try:
        contury_vec = model.wv[contury_name]
        capital_vec = model.wv[capital_name]
        
        relation_vec = contury_vec - capital_vec
        relation_vec_list.append(relation_vec)

    except:
        continue
```

```python
# 任意の関係ベクトルがすべての国ベクトルの出力にきくか

top_list = []
count = 0
c_sum = 0
ok_list = []

average_relation_vec = relation_vec[0]

for l in con_list:
    country_name = l[0]
    capital_name = l[1]
    
    try:
        capital_vec = model.wv[capital_name]
        may_country_vec = capital_vec + average_relation_vec
    
        most_similar = model.wv.most_similar([may_country_vec], [], 5) #most_similarに直接ベクトルを入力
        
        top_list = most_similar
        
        for kouho_capital in top_list:
            if kouho_capital[0] == capital_name:
                print("一致")
                print(kouho_capital[0])
                count += 1
                ok_list.append(country_name)
                break
    
        print(country_name + " : " + capital_name)
        print(most_similar)
        print("\n")
        c_sum += 1

    except:
        print("エラー")
        print(country_name + " : " + capital_name)
        continue
print((count/c_sum)*100)
```

# 関係ベクトルと関係ベクトルの類似度を計算する


```python
model.wv.most_similar([relation_vec_list[0], relation_vec_list[1]], [], 5)
```

```python
import numpy as np

def cal_cos_similarity(q, d):
    return np.dot(q, d) / (np.linalg.norm(q) * np.linalg.norm(d))
```

```python
cal_cos_similarity(relation_vec_list[10], relation_vec_list[1])
```

```python
cal_cos_similarity(relation_vec_list[2], relation_vec_list[1])
```

# 関係ベクトル郡のそれぞれの類似度をマッピングする

```python
len(relation_vec_list)
```

```python
cal_cos_similarity_list = []

for i in range(len(relation_vec_list)):
    if i + 1 >= len(relation_vec_list) :
        break
    similarity = cal_cos_similarity(relation_vec_list[i], relation_vec_list[ i+1])
    cal_cos_similarity_list.append(similarity)
```
