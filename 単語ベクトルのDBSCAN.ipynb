```python
import gensim

model = gensim.models.Word2Vec.load("latest-ja-word2vec-gensim-model/word2vec.gensim.model")
```

```python
# 国と首都のリストを作成

con_list = []

f = open('country.txt', 'r')

datalist = f.readlines()
for i in range(193):
    # print(datalist[i].split("\t"))
    sample = datalist[i].split("\t")
    
    sample[1] = sample[1].rstrip()
    con_list.append(sample)
    
f.close()
```

```python
# 国ベクトル - 首都ベクトル
relation_vec_list = []

for llist in con_list:
    contury_name = str(llist[0])
    capital_name = str(llist[1])
    
    
    # 単語の意味を足し引きして、類似度の高いものを上から１０個表示する
    try:
        contury_vec = model.wv[contury_name]
        capital_vec = model.wv[capital_name]
        
        relation_vec = contury_vec - capital_vec
        relation_vec_list.append(relation_vec)

    except:
        continue
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

print("relation_vec_list数 : ", len(relation_vec_list))

for eps in range(1,3,1):
    eps = eps
    
    for minPts in range(1,20):
        
        dbscan = DBSCAN(eps=eps,min_samples=minPts).fit(relation_vec_list)
        
        print(dbscan)
        
        y_dbscan = dbscan.labels_
        
        print("y_dbscan", y_dbscan)
        
        print("eps:",eps,",minPts:", minPts)
        
        # outlier数
        print("outlier数", len(np.where(y_dbscan ==-1)[0]))
        
        # クラスタ数
        print("クラスタ数", np.max(y_dbscan) +2)
        
        # クラスタ1に含まれる点の数
        print("クラスタ1に含まれる点の数", len(np.where(y_dbscan ==0)[0]))
        
        # クラスタ2に含まれる点の数
        print("クラスタ2に含まれる点の数", len(np.where(y_dbscan ==1)[0]))
        
        # クラスタ3に含まれる点の数
        print("クラスタ3に含まれる点の数", len(np.where(y_dbscan ==2)[0]))
        
        print("                            ")
```

```python
import MeCab
import re

mecab = MeCab.Tagger ()
pattern = re.compile('[a-zA-Z]')
noun_list = []

i = 0

for text in model.wv.index2word:
    #print("text", text)
        
    mecab.parse('')#文字列がGCされるのを防ぐ
    node = mecab.parseToNode(text)
    
    b = re.search(pattern,text)  
    #print("result", result)
        
    if b: 
        #print("英語")
        #print(result.group())
        continue 
    
    i +=1
    
    while node:
        #単語を取得
        word = node.surface
        
        #品詞を取得
        pos = node.feature.split(",")[1]
        #print('{0} , {1}'.format(word, pos))
        
        #print("pos", pos)
        #if pos == "一般" or pos == "固有名詞":
        if pos == "固有名詞":
            noun_list.append(word)
        
        #次の単語に進める
        node = node.next
        #print("while")
```

# 単語ベクトル集合を得る

```python
import pickle

f = open('koyuu_noun_dic.binaryfile','rb')
noun_dic= pickle.load(f)
```

```python
import numpy as np

noun_vec_list = list(noun_dic.values())

re_noun_vec_list = []
```

```python
import random 

sample_list = random.sample(noun_vec_list, int(len(noun_vec_list)/2))
```

```python
len(sample_list)
```

```python
for i in sample_list:
    a = i.reshape(50,)
    re_noun_vec_list.append(a)

re_noun_vec_list = np.array(re_noun_vec_list)
```

```python
len(re_noun_vec_list)
```

```python
import pickle

f = open('sample_noun_vec_list.binaryfile','wb')
#f = open('koyuu_noun_sample_dic.binaryfile','wb')
pickle.dump(re_noun_vec_list, f)
f.close
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

eps_minPts = {}
minPts_data = {}

print("noun_vec_list数 : ", len(re_noun_vec_list))

for eps in range(5,50,5):
    eps = eps/10
    
    for minPts in range(1,100, 10):
        
        dbscan = DBSCAN(eps=eps,min_samples=minPts).fit(re_noun_vec_list)
        
        #print(dbscan)
        
        y_dbscan = dbscan.labels_
        
        print("y_dbscan", y_dbscan)
        
        print("eps:",eps,",minPts:", minPts)
        
        # outlier数
        print("outlier数", len(np.where(y_dbscan ==-1)[0]))
        
        # クラスタ数
        print("クラスタ数", np.max(y_dbscan) +2)
        
        # クラスタ1に含まれる点の数
        print("クラスタ1に含まれる点の数", len(np.where(y_dbscan ==0)[0]))
        
        # クラスタ2に含まれる点の数
        print("クラスタ2に含まれる点の数", len(np.where(y_dbscan ==1)[0]))
        
        # クラスタ3に含まれる点の数
        print("クラスタ3に含まれる点の数", len(np.where(y_dbscan ==2)[0]))
        
        print("                            ")
        
        minPts_data[minPts] = y_dbscan
        eps_minPts[eps] = minPts_data
    print(eps) 
    break
```

```python
import pickle

f = open('dbscan_dic.binaryfile','wb')
pickle.dump(eps_minPts,f)
f.close
```
